```python
def convert_k_to_num(val):
    if pd.isna(val):
        return 0  # Or np.nan if you prefer to handle nulls later
    
    val = str(val).lower().strip()
    
    if 'k' in val:
        # Remove 'k' and multiply by 1000
        return float(val.replace('k', '')) * 1000
    else:
        # Just convert the existing string number to float
        return int(val)

# Apply to your column
df_coursera['reviewcount'] = df_coursera['reviewcount'].apply(convert_k_to_num)

# Verify the result
print(df_coursera['reviewcount'].head())
```

```python
import pandas as pd
import re
import numpy as np

def standardize_to_months(text):
    if pd.isna(text): return np.nan
    text = str(text).lower()
    
    # 1. Standardize synonyms and languages
    text = text.replace('meses', 'months').replace('mes', 'month').replace('a', '-')
    
    # 2. Extract all numbers (handles decimals and ranges)
    nums = [float(n) for n in re.findall(r"[-+]?\d*\.\d+|\d+", text)]
    if not nums: return np.nan
    
    # 3. Get the average if it's a range (e.g., 3-6 becomes 4.5)
    val = sum(nums) / len(nums)
    
    # 4. Convert units to months
    if 'year' in text:
        return val * 12
    elif 'week' in text:
        return val / 4.33 # Average weeks in a month
    elif 'hour' in text:
        return (val / 40) # Assuming 40 hours of study = 1 month
    else:
        return val # Already in months

# Apply the function
df_coursera['duration_months'] = df_coursera['duration'].apply(standardize_to_months)
```
